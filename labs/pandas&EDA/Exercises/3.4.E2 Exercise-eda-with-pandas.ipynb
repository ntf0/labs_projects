# Scenario

You're an analyst working for CoolBricks - a new toy company who wants to compete with LEGO by producing their own bricks and sets. To better understand what kind of sets to create and at what price points, you are tasked with analyzing LEGO's catalogue over time.

The data comes from Kaggle: [https://www.kaggle.com/datasets/alexracape/lego-sets-and-prices-over-time](https://www.kaggle.com/datasets/alexracape/lego-sets-and-prices-over-time)

We are going to use `pandas` to read and explore this data.

### Business question

What is the biggest set we could build (in terms of number of pieces) and what sort of price would we want to charge?

#### 1. Import `pandas` and read in the LEGO dataset

```python
import pandas as pd
```

#### 2. So we can focus on just the columns that help answer the question, select only the following columns into a new `DataFrame` variable:

- name
- theme
- number of pieces
- retail price (USD_MSRP)

```python
df = pd.read_csv('./data/lego.csv', encoding='latin-1')
DataFrame = df[['Name', 'Theme', 'Pieces', 'USD_MSRP']]
```

#### 3. How many sets have a price greater than $100?

```python
DataFrame[DataFrame['USD_MSRP'] > 100]['USD_MSRP'].count()
```

#### 4. What are the top 10 sets with the highest number of pieces?

```python
DataFrame.sort_values(by='Pieces', ascending=False).head(10)
```

#### 5. Create a new column to measure "price per piece", which should be the price (USD_MSRP) divided by the number of pieces

```python
# .copy() creates an independent copy of the filtered DataFrame,
# so changes to valid_sets will not affect the original DataFrame.
# cleaning the data for doesn't print 'inf' value in Price_per_Piece
valid_sets = DataFrame[DataFrame['Pieces'] > 0].copy()

valid_sets['Price_per_Piece'] = valid_sets['USD_MSRP'] / valid_sets['Pieces']

# Add the new colmuns with valid values 
DataFrame['Price_per_Piece'] = valid_sets['Price_per_Piece']
```

#### 6. Use this column to find the top 10 most expensive sets by "price per piece"

```python
DataFrame.sort_values(by='Price_per_Piece', ascending=False).head(10)
```

#### 7. Narrow the search to show the top 10 most expensive sets by "price per piece" for the "Icons" theme *only*

```python
DataFrame[DataFrame['Theme'] == 'Icons'].sort_values(by='Price_per_Piece', ascending=False).head(10)
```

#### 8. What are the top 10 cheapest sets by "price per piece" in the "Icons" theme?

```python
DataFrame[DataFrame['Theme'] == 'Icons'].sort_values(by='Price_per_Piece').head(10)
```

#### 9. Export the `DataFrame` you created in question 2. to a new CSV file.

```python
DataFrame.to_csv('lego_filtered.csv', index=False)
```
